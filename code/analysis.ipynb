```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['figure.figsize'] = [15, 9]
```

```python
npzfile = np.load("../data/metrics/loss-12-6-17:53.npz")
loc_loss, class_loss, overall_loss = npzfile[npzfile.files[0]], npzfile[npzfile.files[1]], npzfile[npzfile.files[2]]
```

```python
x = [i for i in range(len(loc_loss))]
plt.plot(x, loc_loss, label='localization')
plt.plot(x, class_loss, label='classification')
plt.plot(x, overall_loss, label='total')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title(f"Initial loss: {overall_loss[0]:.3f}; Final loss: {overall_loss[-1]:.3f}")

plt.legend()
plt.savefig('model-loss.png')
```
